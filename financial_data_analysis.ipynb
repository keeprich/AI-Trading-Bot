# Loading financial data from the web

```python
import yfinance as yf
import pandas as pd
```

```python
start ="2014-10-01"
end = "2022-10-01"
```

```python
symbol = "EURUSD=X"
```

# Example tickets

- BA: Boeing (US stocks)
- MSFT: Microsoft Cort (US Stock)
- ^DJI: Dow Johns Industrial Average (US Stock index)
- EURUSD=X
- APPL - Apple shares
- GC=F: Gold price
- BTC-USD: bitcoin in USD

```python
yf.download(symbol, start, end)
```

```python
df = yf.download(symbol, start, end)
```

```python
df.info()
```

```python
# Save data

df.to_csv("EURUSD.csv")
```
