```python
# Create Buy/Sell signals based on conditions
data['BuySignal'] = (data['RSI'] < 30) & (data['c'] <= data['LowerBand']) & (data['SMA50'] > data['SMA200'])
data['SellSignal'] = (data['RSI'] > 70) & (data['c'] >= data['UpperBand']) & (data['SMA50'] < data['SMA200'])

# Display signals (True means signal is triggered)
print(data[['c', 'RSI', 'UpperBand', 'LowerBand', 'SMA50', 'SMA200', 'BuySignal', 'SellSignal']].tail(10))
```

# Interpretation of Results

## Bollinger Bands:
When the price hits the upper band, it could be an overbought condition (potential sell). When it hits the lower band, it might indicate an oversold condition (potential buy).


## RSI:
An RSI above 70 is considered overbought (potential sell), and an RSI below 30 is considered oversold (potential buy).


## SMA Crossovers:
A Golden Cross (50-period SMA crosses above 200-period SMA) is a bullish signal, while a Death Cross (50-period SMA crosses below 200-period SMA) is bearish.

```python
import tpqoa
import pandas as pd
import numpy as np

# Initialize OANDA API connection
api = tpqoa.tpqoa("oanda.cfg")

# Fetch historical M1 data for EUR/USD from 2020 to present
data = api.get_history(
    instrument="EUR_USD",
    start="2020-01-01",   # Start date
    end="2023-11-01",     # End date (adjust as needed)
    granularity="M1",               # 1-minute granularity
    price="M"                       # "M" for mid prices
)

# Calculate RSI and simulate trades based on it
RSI_WINDOW = 14                 # RSI lookback period
RSI_BUY_THRESHOLD = 30          # RSI threshold for buying
RSI_SELL_THRESHOLD = 70         # RSI threshold for selling
UNITS = 1000                    # Number of units to buy/sell per trade

# Calculate RSI using the close prices
data['Price_Change'] = data['c'].diff()  # Calculate price changes
data['Gain'] = np.where(data['Price_Change'] > 0, data['Price_Change'], 0)
data['Loss'] = np.where(data['Price_Change'] < 0, -data['Price_Change'], 0)

# Calculate rolling average of gains and losses
data['Avg_Gain'] = data['Gain'].rolling(window=RSI_WINDOW).mean()
data['Avg_Loss'] = data['Loss'].rolling(window=RSI_WINDOW).mean()

# Calculate RSI
data['RS'] = data['Avg_Gain'] / data['Avg_Loss']
data['RSI'] = 100 - (100 / (1 + data['RS']))

# Simulate Buy/Sell Signals
data['Position'] = 0  # 1 for buy, -1 for sell, 0 for hold
data['Profit'] = 0    # Track profit on each trade

# Iterate over data to make decisions based on RSI
for i in range(RSI_WINDOW, len(data)):
    if data['RSI'].iloc[i] < RSI_BUY_THRESHOLD and data['Position'].iloc[i-1] != 1:
        data.at[data.index[i], 'Position'] = 1  # Buy
        print(f"Buy at {data['c'].iloc[i]} on {data.index[i]}")
    elif data['RSI'].iloc[i] > RSI_SELL_THRESHOLD and data['Position'].iloc[i-1] != -1:
        data.at[data.index[i], 'Position'] = -1  # Sell
        print(f"Sell at {data['c'].iloc[i]} on {data.index[i]}")

# Calculate Profit/Loss for each position change
data['Price_Change'] = data['c'].diff()  # Price change between each row
data['Profit'] = data['Price_Change'] * UNITS * data['Position'].shift()

# Total profit from all trades
total_profit = data['Profit'].sum()
print(f"Total simulated profit: {total_profit}")

# Plotting RSI and buy/sell points for visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))

# Plot closing price
plt.subplot(2, 1, 1)
plt.plot(data.index, data['c'], label="Close Price", color='blue')
plt.plot(data[data['Position'] == 1].index, data['c'][data['Position'] == 1], '^', markersize=10, color='g', label='Buy Signal')
plt.plot(data[data['Position'] == -1].index, data['c'][data['Position'] == -1], 'v', markersize=10, color='r', label='Sell Signal')
plt.title('EUR/USD Close Price with Buy/Sell Signals')
plt.legend()

# Plot RSI
plt.subplot(2, 1, 2)
plt.plot(data.index, data['RSI'], label="RSI", color='purple')
plt.axhline(RSI_BUY_THRESHOLD, linestyle='--', color='green', label='Buy Threshold (30)')
plt.axhline(RSI_SELL_THRESHOLD, linestyle='--', color='red', label='Sell Threshold (70)')
plt.title('Relative Strength Index (RSI)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import tpqoa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Initialize OANDA API connection
api = tpqoa.tpqoa("oanda.cfg")

# Fetch historical M1 data for EUR/USD from 2020 to present
data = api.get_history(
    instrument="EUR_USD",
    start="2020-01-01",   # Start date
    end="2023-11-01",     # End date (adjust as needed)
    granularity="M1",               # 1-minute granularity
    price="M"                       # "M" for mid prices
)
```

```python


# Parameters for the strategy
RSI_WINDOW = 14                 # RSI lookback period
RSI_BUY_THRESHOLD = 30          # RSI threshold for buying
RSI_SELL_THRESHOLD = 70         # RSI threshold for selling
UNITS = 1000                    # Number of units to buy/sell per trade
current_position = 0            # Tracks if we are currently in a position (1 = long, -1 = short, 0 = no position)

# Calculate RSI using the close prices
data['Price_Change'] = data['c'].diff()  # Calculate price changes
data['Gain'] = np.where(data['Price_Change'] > 0, data['Price_Change'], 0)
data['Loss'] = np.where(data['Price_Change'] < 0, -data['Price_Change'], 0)

# Calculate rolling average of gains and losses
data['Avg_Gain'] = data['Gain'].rolling(window=RSI_WINDOW).mean()
data['Avg_Loss'] = data['Loss'].rolling(window=RSI_WINDOW).mean()

# Calculate RSI
data['RS'] = data['Avg_Gain'] / data['Avg_Loss']
data['RSI'] = 100 - (100 / (1 + data['RS']))

# Initialize lists to track orders
buy_signals = []
sell_signals = []

# Iterate over data to make decisions based on RSI
for i in range(RSI_WINDOW, len(data)):
    current_rsi = data['RSI'].iloc[i]
    close_price = data['c'].iloc[i]

    # Buy Signal
    if current_rsi < RSI_BUY_THRESHOLD and current_position != 1:
        api.create_order(instrument="EUR_USD", units=UNITS, type="MARKET")  # Place buy order
        buy_signals.append((data.index[i], close_price))  # Record buy signal
        current_position = 1  # Update current position
        print(f"Buy order placed at {close_price} on {data.index[i]} (RSI: {current_rsi:.2f})")

    # Sell Signal
    elif current_rsi > RSI_SELL_THRESHOLD and current_position != -1:
        api.create_order(instrument="EUR_USD", units=-UNITS, type="MARKET")  # Place sell order
        sell_signals.append((data.index[i], close_price))  # Record sell signal
        current_position = -1  # Update current position
        print(f"Sell order placed at {close_price} on {data.index[i]} (RSI: {current_rsi:.2f})")

# Plot RSI and buy/sell signals
plt.figure(figsize=(14, 7))

# Plot closing price with buy/sell signals
plt.subplot(2, 1, 1)
plt.plot(data.index, data['c'], label="Close Price", color='blue')
plt.scatter(*zip(*buy_signals), color='green', marker='^', label='Buy Signal')
plt.scatter(*zip(*sell_signals), color='red', marker='v', label='Sell Signal')
plt.title('EUR/USD Close Price with Buy/Sell Signals')
plt.legend()

# Plot RSI
plt.subplot(2, 1, 2)
plt.plot(data.index, data['RSI'], label="RSI", color='purple')
plt.axhline(RSI_BUY_THRESHOLD, linestyle='--', color='green', label='Buy Threshold (30)')
plt.axhline(RSI_SELL_THRESHOLD, linestyle='--', color='red', label='Sell Threshold (70)')
plt.title('Relative Strength Index (RSI)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from datetime import datetime, timedelta

# Define date range for data retrieval
start_date = datetime(2020, 1, 1)
end_date = datetime.now()  # Fetch data up to the current date

# Initialize empty DataFrame to store data
all_data = pd.DataFrame()

# Loop to fetch data in monthly chunks
while start_date < end_date:
    # Set chunk end date (1 month after start_date or end_date, whichever is earlier)
    chunk_end_date = min(start_date + timedelta(days=30), end_date)

    # Fetch data for this date range
    data_chunk = api.get_history(
        instrument="EUR_USD",
        start=start_date.isoformat(),
        end=chunk_end_date.isoformat(),
        granularity="M1",
        price="M"
    )

    # Append chunk to all_data DataFrame
    all_data = pd.concat([all_data, data_chunk])

    # Move to next month
    start_date = chunk_end_date

# Save the final dataset to a CSV file for future use
all_data.to_csv("EURUSD_M1_2020_2023.csv", index=False)
```

```python
import yfinance as yf
import pandas_ta as ta
import matplotlib.pyplot as plt

# Download data
df = yf.download('EURUSD=X', start='2014-10-01', end='2022-09-30', interval='1d')

# Check column names and structure
print(df.columns)

# Rename columns to simplify access
df.columns = df.columns.droplevel()  # Drop the multi-level column names (e.g., 'Adj Close', 'Close', etc.)

# Calculate RSI (Relative Strength Index)
df['RSI'] = ta.rsi(df['Close'], length=14)

# Calculate 50-period Simple Moving Average (SMA)
df['SMA_50'] = ta.sma(df['Close'], length=50)

# Calculate Bollinger Bands
bbands = ta.bbands(df['Close'], length=20)
df['Upper_Band'] = bbands['BBU_20_2.0']
df['Lower_Band'] = bbands['BBL_20_2.0']

# Calculate MACD
df['MACD'], df['MACD_Signal'], df['MACD_Hist'] = ta.macd(df['Close'], fast=12, slow=26, signal=9)

# Display last few rows to confirm indicators are computed
print(df.tail())

# Plot the technical indicators

# Create subplots
fig, axs = plt.subplots(4, 1, figsize=(12, 12), sharex=True)

# Plot 1: Closing price and SMA_50
axs[0].plot(df.index, df['Close'], label='Close Price', color='blue')
axs[0].plot(df.index, df['SMA_50'], label='50-period SMA', color='orange')
axs[0].set_title('Close Price with 50-period SMA')
axs[0].legend()

# Plot 2: RSI with overbought/oversold levels
axs[1].plot(df.index, df['RSI'], label='RSI', color='purple')
axs[1].axhline(30, color='red', linestyle='--', label='Oversold (30)')
axs[1].axhline(70, color='green', linestyle='--', label='Overbought (70)')
axs[1].set_title('RSI (14)')
axs[1].legend()

# Plot 3: Bollinger Bands
axs[2].plot(df.index, df['Close'], label='Close Price', color='blue')
axs[2].plot(df.index, df['Upper_Band'], label='Upper Bollinger Band', color='red')
axs[2].plot(df.index, df['Lower_Band'], label='Lower Bollinger Band', color='green')
axs[2].set_title('Bollinger Bands (20,2)')
axs[2].legend()

# Plot 4: MACD and Signal Line
axs[3].plot(df.index, df['MACD'], label='MACD', color='blue')
axs[3].plot(df.index, df['MACD_Signal'], label='MACD Signal', color='orange')
axs[3].bar(df.index, df['MACD_Hist'], label='MACD Histogram', color='gray', alpha=0.3)
axs[3].set_title('MACD (12,26,9)')
axs[3].legend()

# Show the plots
plt.tight_layout()
plt.show()
```
