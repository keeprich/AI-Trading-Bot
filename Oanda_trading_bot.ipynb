```python
import time
import pandas as pd
import tpqoa

# OANDA Account details

INSTRUMENT = 'EUR_USD'  # Replace with your chosen currency pair

# Initialize the tpqoa API client
api = tpqoa.tpqoa("oanda.cfg")

# Define moving average periods
SHORT_MA_PERIOD = 5
LONG_MA_PERIOD = 20
TRADE_UNITS = 100  # Size of each trade in units
STOP_LOSS = 0.02  # 2% stop loss
TAKE_PROFIT = 0.03  # 3% take profit

# Fetch historical data and calculate moving averages
def fetch_data():
    print("Fetching data...")
    # Hard-coded start and end time in ISO format
    start_time = "2024-11-01"  # Adjust this date as needed
    end_time = "2024-11-07"    # Adjust this date as needed

    # Fetch historical data using tpqoa's get_history()
    data = api.get_history(
        instrument=INSTRUMENT,
        start=start_time,
        end=end_time,
        granularity='M5',
        price='M'
    )

    print("Data fetched successfully.")
    prices = data['c']  # Get the closing prices
    return pd.Series(prices)

def calculate_moving_averages(prices):
    print("Calculating moving averages...")
    sma_short = prices.rolling(window=SHORT_MA_PERIOD).mean().iloc[-1]
    sma_long = prices.rolling(window=LONG_MA_PERIOD).mean().iloc[-1]
    print(f"Short MA: {sma_short}, Long MA: {sma_long}")
    return sma_short, sma_long

def place_trade(side):
    """Place a market order with tpqoa"""
    try:
        if side == 'buy':
            api.create_order(
                instrument=INSTRUMENT,
                units=TRADE_UNITS,
                suppress=False,  # Set to False to show order response
                stop_loss=STOP_LOSS,
                take_profit=TAKE_PROFIT
            )
            print("Buy order placed successfully.")
        elif side == 'sell':
            api.create_order(
                instrument=INSTRUMENT,
                units=-TRADE_UNITS,
                suppress=False,  # Set to False to show order response
                stop_loss=STOP_LOSS,
                take_profit=TAKE_PROFIT
            )
            print("Sell order placed successfully.")
    except Exception as e:
        print(f"Error placing {side} order: {e}")

def trading_logic():
    print("Starting trading logic check...")
    prices = fetch_data()
    sma_short, sma_long = calculate_moving_averages(prices)

    if sma_short > sma_long:
        print("Crossover up detected - Placing Buy Order")
        place_trade('buy')
    elif sma_short < sma_long:
        print("Crossover down detected - Placing Sell Order")
        place_trade('sell')
    else:
        print("No crossover - No trade action")

# Main trading loop
while True:
    try:
        trading_logic()
        time.sleep(300)  # Wait 5 minutes (300 seconds) between checks
    except Exception as e:
        print(f"Error in trading loop: {e}")
```
